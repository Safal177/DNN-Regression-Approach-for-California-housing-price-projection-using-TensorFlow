#### Separate the data into training and testing subsets

```python
from sklearn.model_selection import train_test_split

train_data, test_data, train_labels, test_labels = train_test_split(data[california_housing_data.feature_names], data['target'], test_size = 0.25, random_state = 42)
```

```python
print(train_data.shape  )
```

```python
print(test_data.shape  )
```

#### Apply standardization (Normalization) to the feature variables

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard_scaler = StandardScaler()

train_features = standard_scaler.fit_transform(train_data)

test_features = standard_scaler.transform(test_data)

print(train_features)

print(":\n")

print(test_features)
```

### Creating the Deep Neural Network (DNN) configuration with TensorFlow/Keras

```python
from tensorflow.keras.layers import Dense
```

```python
#  Make a simple deep neural network with a hidden layer of 20 neurons between the input layer (8 features) and the output layer (1 neuron) 
#  including optimizer. 


def deep_neural_network_model():
    dnn_model = keras.Sequential([Dense(20, activation = tf.nn.relu, input_shape = [train_features.shape[1]]),
                                  Dense(1)])

    dnn_model.compile(optimizer = tf.optimizers.Adam(),
                      loss = 'mae',
                      metrics = ['mean_absolute_error'])
    return dnn_model
```

### Deep learning model training and progress tracking

```python
from tensorflow import keras
```

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
```

```python
model = deep_neural_network_model()

early_stop_model = keras.callbacks.EarlyStopping(monitor = 'val_loss', patience = 20)


history = model.fit(train_features, train_labels, epochs = 30, verbose = 0, validation_split = 0.15,
                             callbacks = [early_stop_model])

# Create a Pandas DataFrame from the training history
training_history_dataframe = pd.DataFrame(history.history)
training_history_dataframe['epoch'] = history.epoch

validation_mean_absolute_error = float(training_history_dataframe['val_mean_absolute_error'].iloc[-1])

print()

print(' Mean absolute error calculated on the validation data: {}'.format(round(validation_mean_absolute_error, 2)))
```

```python
# Analyze the trained model using the test features

mae, _ = model.evaluate(test_features, test_labels)
print('Mean absolute error measured on the test data: {}'.format(round(mae, 2)))
```
